Day Wise Total Message Count Calculation ##################################################################

```python
from collections import defaultdict
from firebase_admin import firestore
from datetime import datetime

# Reference to the 'customer-chat-lists' collection
collection_ref = db.collection("customer-chat-lists")

# Stream all documents in the collection
docs = collection_ref.stream()

# Dictionary to store counts grouped by date
date_counts = defaultdict(int)

for doc in docs:
    data = doc.to_dict()
    created_at = data.get("createdAt")

    if isinstance(created_at, datetime):
        # Format the datetime to a date string (e.g., '2025-06-11')
        date_str = created_at.date().isoformat()
        date_counts[date_str] += 1
    else:
        print(f"Document {doc.id} has no valid 'createdAt' timestamp.")

# Print the total count grouped by date
print("\nTotal documents in 'customer-chat-lists' grouped by 'createdAt' date:\n")
for date_str, count in sorted(date_counts.items()):
    print(f"{date_str}: {count}")

```

Customer And Admin Wise Total Messages Calculation ############################################################

```python
from firebase_admin import firestore
from collections import defaultdict
import datetime

# Initialize Firestore (assumes firebase_admin.initialize_app() already called)
db = firestore.client()

# Reference to the 'chats' document
doc_ref = db.collection("chat-updated").document("chats")

# Get all subcollections
subcollections = doc_ref.collections()

# Prepare dictionary to hold counts
admin_counts = defaultdict(int)
customer_counts = defaultdict(int)

# Iterate through subcollections
for subcol in subcollections:
    for subdoc in subcol.stream():
        data = subdoc.to_dict()
        
        # Ensure both 'role' and 'createdAt' exist
        if 'role' in data and 'createdAt' in data:
            role = data['role']
            created_at = data['createdAt']

            # Convert to date object
            if isinstance(created_at, datetime.datetime):
                date_key = created_at.date()

                # Increment counts based on role
                if role == 'Admin':
                    admin_counts[date_key] += 1
                elif role == 'Customer':
                    customer_counts[date_key] += 1

# Display results
print("Day-wise Admin Chat Counts:")
for day, count in sorted(admin_counts.items()):
    print(f"{day}: {count} chats")

print("\nDay-wise Customer Chat Counts:")
for day, count in sorted(customer_counts.items()):
    print(f"{day}: {count} chats")
```

Response Time Calculation ############################################################################

```python
from datetime import datetime
from firebase_admin import firestore
from statistics import mean

db = firestore.client()
doc_ref = db.collection("chat-updated").document("chats")
subcollections = doc_ref.collections()

response_times = []

for subcol in subcollections:
    messages = sorted(
        [doc.to_dict() for doc in subcol.stream()],
        key=lambda x: x.get('createdAt')
    )

    last_customer_time = None

    for msg in messages:
        role = msg.get("role")
        created_at = msg.get("createdAt")

        if isinstance(created_at, datetime):
            if role == "Customer":
                last_customer_time = created_at
            elif role == "Admin" and last_customer_time:
                diff = (created_at - last_customer_time).total_seconds()
                if diff >= 0:
                    response_times.append(diff)
                last_customer_time = None

if response_times:
    print(f"Min. Response Time: {min(response_times):.2f} seconds")
    print(f"Max. Response Time: {max(response_times):.2f} seconds")
    print(f"Average Response Time: {mean(response_times):.2f} seconds")
else:
    print("No response time data available.")
```

Daily Respont Time Calculation #######################################################

```python
from datetime import datetime
from firebase_admin import firestore
from statistics import mean
from collections import defaultdict

db = firestore.client()
doc_ref = db.collection("chat-updated").document("chats")
subcollections = doc_ref.collections()

# Dictionary to hold response times per day
daily_response_times = defaultdict(list)

for subcol in subcollections:
    messages = sorted(
        [doc.to_dict() for doc in subcol.stream()],
        key=lambda x: x.get('createdAt')
    )

    last_customer_time = None

    for msg in messages:
        role = msg.get("role")
        created_at = msg.get("createdAt")

        if isinstance(created_at, datetime):
            if role == "Customer":
                last_customer_time = created_at
            elif role == "Admin" and last_customer_time:
                diff = (created_at - last_customer_time).total_seconds()
                if diff >= 0:
                    day = last_customer_time.date().isoformat()
                    daily_response_times[day].append(diff)
                last_customer_time = None

# Print day-wise min, max, and average
for day in sorted(daily_response_times):
    times = daily_response_times[day]
    print(f"\nDate: {day}")
    print(f"   Min RTime: {min(times):.2f} sec")
    print(f"   Max RTime: {max(times):.2f} sec")
    print(f"   Avg RTime: {mean(times):.2f} sec")
```
